```python
import random
import numpy as np

# Generate random coordinates aligned to the left
num1 = 200
middel1 = int(num1 * 0.6)
num1 = num1 - middel1
num2 = 200
middel2 = int(num2 * 0.6)
num2 = num2 - middel2

num_left_points = num1
left_coordinates = [(random.uniform(0, 7), random.uniform(0, 10)) for _ in range(num_left_points)]

# Generate random coordinates aligned to the right
num_right_points = num2
right_coordinates = [(random.uniform(13, 20), random.uniform(0, 10)) for _ in range(num_right_points)]

num_middle_points1 = middel1
middle_coordinates1 = [(random.uniform(7, 13), random.uniform(0, 10)) for _ in range(num_middle_points1)]
num_middle_points2 = middel2
middle_coordinates2 = [(random.uniform(7, 13), random.uniform(0, 10)) for _ in range(num_middle_points2)]

left_coordinates = np.array(left_coordinates)
middle_coordinates1 = np.array(middle_coordinates1)
middle_coordinates2 = np.array(middle_coordinates2)
right_coordinates = np.array(right_coordinates)
```

```python
import matplotlib.pyplot as plt
plt.scatter(left_coordinates[:,0],left_coordinates[:,1], c='red', label='Group 1')
plt.scatter(middle_coordinates1[:,0], middle_coordinates1[:,1], c='red', label='Group 1')

plt.scatter(right_coordinates[:,0], right_coordinates[:,1], c='green', label='Group 2')
plt.scatter(middle_coordinates2[:,0], middle_coordinates2[:,1], c='green', label='Group 2')


# 添加标题和标签
plt.title('Scatter Plot of 3 Groups')
plt.xlabel('X')
plt.ylabel('Y')

# 添加图例
plt.legend()

# 显示散点图
plt.show()
```

```python
all_datas=[]
col_list = ['ALOX5AP','ALOX5AP','PLA2G10','PLA2G10'];
col_list1 = ['NOD2','ITGA9','SMAD3','GRM6']
not_path = [i for i in not_attrs if(not(i in t))]
tt = not_attrs
np.random.shuffle(tt)
split_list = np.array_split(tt, 4)

for j in range(len(col_list)):
    # rnd_not_attr = random.sample(not_attrs,4000)
    
    dn = adata1.copy().to_df()
    col = col_list[j]
    dn[col] = [dn[col][i] + np.random.randint(1, 4) for i in range(len(dn))]
    
    for k in not_path:
        dn[k] = 0
    
    for k in split_list[j][:4000]:
        dn[k] = [dn[k][i] + np.random.randint(1, 2) for i in range(len(dn))]
        
    n1 = [df['lr_score'].iloc[i][['ligand','receptor']].tolist() for i in range(len(df['lr_score'])) if df['lr_score'].iloc[i]['ligand'] == col_list[j]]
    n2 = [df['lr_score'].iloc[i][['ligand','receptor']].tolist() for i in range(len(df['lr_score'])) if df['lr_score'].iloc[i]['receptor'] == col_list1[j]]
    n1.extend(n2)
    my_set = set(np.array(n1).flatten().tolist())
    my_set.remove(col_list[j])
    my_set.remove(col_list1[j])
    for k in my_set:
        dn[k] = 0 
    datas1 = sc.AnnData(dn)
    datas1.obs = adata1.obs
    datas1.obsm['meta'] =  adata1.obsm['meta']
    
    
    dn = adata2.copy().to_df()
    col = col_list1[j]
    dn[col] = [dn[col][i] + np.random.randint(1, 4) for i in range(len(dn))]
    for k in not_path:
        dn[k] = 0
    for k in split_list[j][:4000]:
        dn[k] = [dn[k][i] + np.random.randint(1, 2) for i in range(len(dn))]

    n1 = [df['lr_score'].iloc[i][['ligand','receptor']].tolist() for i in range(len(df['lr_score'])) if df['lr_score'].iloc[i]['ligand'] == col_list[j]]
    n2 = [df['lr_score'].iloc[i][['ligand','receptor']].tolist() for i in range(len(df['lr_score'])) if df['lr_score'].iloc[i]['receptor'] == col_list1[j]]
    n1.extend(n2)
    my_set = set(np.array(n1).flatten().tolist())
    my_set.remove(col_list[j])
    my_set.remove(col_list1[j])
    for k in my_set:
        dn[k] = 0 
        
    datas2 = sc.AnnData(dn)
    datas2.obs = adata2.obs
    datas2.obsm['meta'] =  adata2.obsm['meta']
    
    tmp = datas1.concatenate(datas2)
    tmp.obsm['meta'].index.name = 'cell'

    tmp.write_h5ad('./datas/{}/AN.h5ad'.format(str(j+1)))
    print(j+1)
    all_datas.append(tmp)
```
