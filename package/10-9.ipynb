```python
file_path = './datas/pre3/sample2'
adata = sc.read(file_path+'.h5')

# 获取每个聚类标签的计数
label_counts = adata.obs['celltype'].value_counts()

# 设置目标每个聚类标签的数据数量（假设为最小计数）
target_count = label_counts.min()

# 存储抽样后的数据
sampled_data = []

# 对每个聚类标签进行遍历
for label in label_counts.index:
    # 获取该聚类标签对应的索引
    label_indices = np.where(adata.obs['celltype'] == label)[0]
    
    # 随机选择目标数量的索引
    selected_indices = np.random.choice(label_indices, size=target_count, replace=False)
    
    # 提取选定的数据
    selected_data = adata[selected_indices]
    
    # 将选定的数据添加到抽样后的数据集中
    sampled_data.append(selected_data)

# 合并抽样后的数据
sampled_data = sampled_data[0].concatenate(sampled_data[1:])
sampled_data.write_h5ad(file_path+'_1.h5')
```
